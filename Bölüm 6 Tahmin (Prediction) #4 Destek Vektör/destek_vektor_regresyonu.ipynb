```python
import pandas as pd

data = pd.read_csv("maaslar.csv")
df = data.copy()
df.head()
```

```python
X = df[["Egitim Seviyesi"]]
Y = df[["maas"]]
```

```python
#svr  kullanrken standard scaler kullanmaz gerekiyor
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
ss = StandardScaler()
x_1 = ss.fit_transform(X)
ss2 = StandardScaler()
y_1 = np.ravel(ss2.fit_transform(Y))
```

```python
y_1
```

```python
from sklearn.svm import SVR
import matplotlib.pyplot as plt
svr = SVR(kernel="rbf")
model = svr.fit(x_1,y_1)
y_pred = model.predict(x_1)
```

```python
plt.scatter(x_1,y_1)
plt.plot(x_1,y_pred)
plt.show()
```
